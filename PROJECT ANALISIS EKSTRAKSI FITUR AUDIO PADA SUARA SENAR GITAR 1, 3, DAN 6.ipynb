```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from scipy.io.wavfile import read
from IPython.display import Audio
from scipy.fft import fft, fftfreq
import numpy as np
import librosa
import antropy as an
import librosa.display
import sklearn
from scipy.stats import entropy as en
```

### FFT

### Analisis dan plot koefisisen FFT senar 1

```python
n1 = dsen1.size
yf1 = fft(dsen1)
xf1 = fftfreq(n1, 1/fs)
plt.figure(figsize=(20,12))
plt.plot(xf1, np.abs(yf1))
plt.xlim(0,3000)
plt.title('Plot FFT senar 1',size = 20)
plt.xlabel('Frequency (Hz)',size = 20);
```

###  Simpan koef senar 1 dalam file .csv

```python
koef_senar_1 = pd.DataFrame((abs(yf1)))
koef_senar_1.to_csv("FFT_suara_senar_1.csv")
koef_senar_1=pd.read_csv('FFT_suara_senar_1.csv')
koef_senar_1
```

```python
n3 = dsen3.size
yf3 = fft(dsen3)
xf3 = fftfreq(n3, 1/fs)
plt.figure(figsize=(20,12))
plt.plot(xf3, np.abs(yf3))
plt.xlim(0,3000)
plt.title('Plot FFT senar 3',size = 20)
plt.xlabel('Frequency (Hz)',size = 20);
```

```python
n6 = dsen6.size
yf6 = fft(dsen6)
xf6 = fftfreq(n6, 1/fs)
plt.figure(figsize=(20,12))
plt.plot(xf6, np.abs(yf6))
plt.xlim(0,3000)
plt.title('Plot FFT senar 6',size = 20)
plt.xlabel('Frequency (Hz)',size = 20);
plt.ylabel('Power Spectral Density / Power (dB)',size = 20)
```

### Energy

### Energy pada senar 1

```python
hop_length = 2207
frame_length = 4414
energy1 = np.array([sum(abs(dsen1[i:i+frame_length]**2))for i in range(0, len(dsen1), hop_length)])
frames1 = range(len(energy1))
t1 = librosa.frames_to_time(frames1, sr=fs, hop_length=hop_length)
plt.figure(figsize=(12,7))
plt.title('Energi pada gelombang senar 1',fontsize=17)
plt.xlabel('Time(s)',fontsize=17)
plt.ylabel('Energy',fontsize=17)
plt.grid(True)
plt.plot(t1,energy1)
plt.show()
```

### Spectral Flux

### Senar 1

```python
spectral_flux_s1 = 0
signal_s1 = dsen1/sum(dsen1)
eps_s1 = 2.2204e-16
for i in (0,dsen1):
    signal_prev_s1 = (i-1)/sum((dsen1-1)+eps_s1)
    spectral_flux_s1 += ((signal_s1-signal_prev_s1)**2)

plt.figure(figsize = (14,6))
plt.plot(spectral_flux_s1)
plt.title('Spectral Flux Senar 1')
```

```python
spectral_flux_1 = pd.DataFrame((spectral_flux_s1))
spectral_flux_1.to_csv("fitur_SF_suara_senar_1.csv")
spectral_flux_1=pd.read_csv('fitur_SF_suara_senar_1.csv')
spectral_flux_1
```

### Senar 3
